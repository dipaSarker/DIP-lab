```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 

# Function to decrease intensity levels
def decrease_intensity_levels(image, bits):
    levels = 2 ** bits
    factor = 256 // levels
    return (image // factor) * factor

def main():
    # Load a grayscale image
    image_path = "img2.jpg"  # Replace with your grayscale image path
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    if image is None:
        print("Error: Image not found. Make sure the path is correct.")
        return

    # Resize to 512x512 if needed
    image = cv2.resize(image, (512, 512))

    # Create a figure to display results
    plt.figure(figsize=(10, 10))

    for i in range(8, 0, -1):
        # Reduce intensity resolution by one bit
        reduced_image = decrease_intensity_levels(image, i)

        # Display the result
        plt.subplot(3, 3, 9 - i)
        plt.imshow(reduced_image, cmap='gray')
        plt.title(f"{i}-bit Resolution")
        plt.axis('off')

    # Display the original image as reference
    plt.subplot(3, 3, 9)
    plt.imshow(image, cmap='gray')
    plt.title("Original Image")
    plt.axis('off')

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```
