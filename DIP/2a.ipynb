```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the grayscale image
image_path = 'img1.jpg'  # Replace with your image path
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Define the range of gray levels to enhance
lower_bound = 100  # Lower bound of gray level
upper_bound = 200  # Upper bound of gray level
brightness_increase = 50  # Amount to increase brightness

# Create a copy of the original image to modify
enhanced_image = image.copy()

# Enhance brightness for the specified range
enhanced_image[(image >= lower_bound) & (image <= upper_bound)] += brightness_increase

# Display original and enhanced images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(enhanced_image, cmap='gray')
plt.title('Brightness Enhanced Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
