```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the grayscale image
image_path = 'img1.jpg'  # Replace with your image path
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Ensure the image is 512x512
image = cv2.resize(image, (512, 512))

# Power Law Transformation (Gamma Correction)
def power_law_transform(image, gamma):
    # Apply power law transformation
    return np.clip(255 * (image / 255) ** gamma, 0, 255).astype(np.uint8)

# Inverse Logarithmic Transformation
def inverse_log_transform(image):
    # Apply inverse logarithmic transformation
    return np.clip(255 * (np.log1p(image / 255)), 0, 255).astype(np.uint8)

# Parameters for transformations
gamma_value = 0.6

# Apply transformations
power_law_result = power_law_transform(image, gamma_value)
inverse_log_result = inverse_log_transform(image)

# Display original and transformed images
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(power_law_result, cmap='gray')
plt.title(f'Power Law Transform (gamma={gamma_value})')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(inverse_log_result, cmap='gray')
plt.title('Inverse Log Transform')
plt.axis('off')

plt.tight_layout()
plt.show()
```
