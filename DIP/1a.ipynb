```python
import cv2
import matplotlib.pyplot as plt

# Load a grayscale image of size 512x512
image_path = "img1.jpg"  # Replace with the path to your image
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Display images with decreasing resolution in a grid layout
plt.figure(figsize=(8, 8))
rows, cols = 3, 3  # Arrange images in 2x2 grid
for i in range(6):  # Reduce resolution 4 times
    resized_image = cv2.resize(image, (512 // (2 ** i), 512 // (2 ** i)))
    resized_image = cv2.resize(resized_image, (512, 512), interpolation=cv2.INTER_NEAREST)
    plt.subplot(rows, cols, i + 1)
    plt.imshow(resized_image, cmap='gray')
    plt.title(f"Level {i}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```
